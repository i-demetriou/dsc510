```python
df = pd.read_csv("salaries.csv")
# get numeric columns
df_numeric = df[['phd', 'service', 'salary']]
# get mean values: the same as df_numeric.mean()
df_numeric.agg('mean')
```

```python
# multiple aggregations for all columns
df_numeric.agg(['max','mean','std'])
```

```python
# multiple statistics on specific columns
df_numeric[['service','salary']].agg(['max','mean','std'])
```

```python
# multiple different statistics for each selected column(s)
df_numeric.agg({'phd': ['max','mean','std'], 'service': ['count', 'mean']})
```

```python
# multiple statistics for each professor rank (for all columns)
df.groupby('rank').agg(['max','mean','std'])
```

```python
#Calculate max, mean, std of the salary column for each professor rank
df.groupby('rank')['salary'].agg(['max','mean','std'])
```

```python
#Calculate max, mean, std of the salary column for each professor rank
df2 = df.groupby('rank').agg({'salary': ['max','mean','std']})
df2
```
