## Section 4: Lists and Slicing

```python
nums = [10, 20, 30, 40, 50]
print(nums[0])     # first element
print(nums[-1])    # last element
print(nums[1:4])   # slicing [20, 30, 40]
print(nums[:3])    # first 3
print(nums[2:])    # from index 2 onwards
```
